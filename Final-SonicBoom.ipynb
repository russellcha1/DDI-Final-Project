```python
import pandas as pd
import requests
import time

url = "https://ll.thespacedevs.com/2.3.0/launches/"

params = {
    "pad__location": 11,
    "mode": "detailed",
    "limit": 100,
    "format": "json"
}

all_results = []

while url:
    response = requests.get(url, params=params)
    response.raise_for_status()

    data = response.json()
    all_results.extend(data["results"])

    url = data["next"]
    params = None

    time.sleep(1)

vandenberg_df = pd.json_normalize(all_results)

vandenberg_df.to_csv("vandenberg_launches.csv", index=False)

vandenberg_df.head()
```

```python
import json
import pandas as pd


def safe_nunique(series):
    """
    Count unique values, including columns containing
    lists or dictionaries.
    """
    def make_hashable(value):
        if isinstance(value, (list, dict)):
            return json.dumps(value, sort_keys=True, default=str)

        return value

    converted = series.map(make_hashable)

    return converted.nunique(dropna=True)


column_profile = pd.DataFrame({
    "dtype": vandenberg_model_df.dtypes.astype(str),

    "non_null_count": (
        vandenberg_model_df
        .notna()
        .sum()
    ),

    "non_null_percent": (
        vandenberg_model_df
        .notna()
        .mean()
        .mul(100)
        .round(2)
    ),

    "unique_values": (
        vandenberg_model_df
        .apply(safe_nunique)
    ),
})

column_profile = column_profile.sort_values(
    "non_null_percent",
    ascending=False,
)

column_profile
```

```python
def clean_psf(col):
    text = (col.astype("string").str.strip().str.upper())
    no_detection = text.str.startswith("ND", na=False)
    numeric_values = pd.to_numeric(text.str.replace(
        "*", "", regex=False), errors="coerce")
    numeric_values = numeric_values.mask(no_detection, 0.0)
    return numeric_values
```

```python
measurement_dataframes = [boom, mantech_before, byu]
for dataframe in measurement_dataframes:
    dataframe["launch_psf"] = clean_psf(
        dataframe["Measured Launch Boom (psf)"])
    dataframe["landing_psf"] = clean_psf(
        dataframe["Measured Landing Boom (psf)"])
```

```python
conus_boom = measurements.loc[measurements['station_latitude'].between(24.4, 49.4) & measurements['station_longitude'].between(
    -125, -66.9) & (measurements['launch_psf'].notna() | measurements['landing_psf'].notna())].copy()
```

```python
boom_threshold = .5
conus_boom['station_launch_boom'] = conus_boom['launch_psf'] >= boom_threshold
conus_boom['station_landing_boom'] = conus_boom['landing_psf'] >= boom_threshold
```

```python
conus_boom['boom_happened'] = (conus_boom['station_launch_boom'].fillna(
    False) | conus_boom['station_landing_boom'].fillna(False)).astype(int)
```

```python
conus_boom['psf_max'] = (
    conus_boom[['launch_psf', 'landing_psf']].max(axis=1, skipna=True))
```

```python
boom_target = (conus_boom.groupby('launch_date', as_index=False).agg(mission=('mission', 'first'), landing=('landing_location', 'first'), max_launch_psf=(
    'launch_psf', 'max'), max_land_psf=('landing_psf', 'max'), max_boom=('psf_max', 'max'), stations=('station_name', 'nunique'), stations_affected=('boom_happened', 'sum'), launch_boom_heard=('launch_psf', 'count'), land_boom_heard=('landing_psf', 'count')))
```

```python
boom_target['conus_boom'] = (
    boom_target['max_boom'] >= boom_threshold).astype(int)
```

```python
launch['launch_datetime_local'] = (pd.to_datetime(
    launch['net'], utc=True, errors='coerce').dt.tz_convert('America/Los_Angeles'))
launch['api_launch_date'] = (
    launch['launch_datetime_local'].dt.tz_localize(None).dt.normalize())
launch['month'] = (launch['launch_datetime_local'].dt.month)
launch['hour'] = (launch['launch_datetime_local'].dt.hour +
                  launch['launch_datetime_local'].dt.minute / 60)
```

```python
launch_features = launch.loc[launch["rocket.configuration.full_name"].astype("string").str.contains("Falcon 9", case=False, na=False), [
    "api_launch_date", "name", "mission.name", "mission.type", "mission.orbit.name", "month", "hour"]].copy()

launch_features = launch_features.rename(
    columns={"name": "api_launch_name", "mission.name": "api_mission_name", "mission.type": "mission_type", "mission.orbit.name": "orbit_name"})

launch_features = (launch_features.sort_values(
    "api_launch_date").drop_duplicates("api_launch_date", keep="first"))
```

```python
model_df = pd.merge_asof(boom_target.sort_values('launch_date'), launch_features.sort_values(
    'api_launch_date'), left_on='launch_date', right_on='api_launch_date', direction='nearest', tolerance=pd.Timedelta(days=1))
```

```python
numeric = ['month', 'hour']
categorical = ['mission_type', 'orbit_name', 'landing']
features = (numeric + categorical)
X = model_df[features].copy()
y = model_df['conus_boom'].astype(int)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=.2, random_state=42, stratify=y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

preprocessor = ColumnTransformer([('num', Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler(
))]), numeric), ('cat', Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('one_hot', OneHotEncoder(handle_unknown='ignore'))]), categorical)])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

log_model = Pipeline([('preprocessor', preprocessor), ('classifier',
                     LogisticRegression(max_iter=2000, random_state=42))])
forest_model = Pipeline([('preprocessor', preprocessor), ('classifier', RandomForestClassifier(
    n_estimators=500, min_samples_leaf=2, random_state=42, n_jobs=-1))])
```

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import StratifiedKFold, cross_val_predict

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

log_param_grid = {'classifier__C': [.03, .1, .3, 1, 3,
                                    10], 'classifier__class_weight': [None, 'balanced']}
forest_param_grid = {'classifier__max_depth': [2, 4, 6, None], 'classifier__min_samples_leaf': [
    1, 2, 4], 'classifier__max_features': ['sqrt', None], 'classifier__class_weight': [None, 'balanced']}

models = {'Logistic regression': log_model, 'Random forest': forest_model}
cv_results = []
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

log_search = GridSearchCV(estimator=log_model, param_grid=log_param_grid,
                          scoring="balanced_accuracy", cv=cv, n_jobs=-1)

log_search.fit(X_train, y_train)

best_log_model = log_search.best_estimator_

print("Best logistic regression parameters:", log_search.best_params_)

print("Best logistic regression CV balanced accuracy:", log_search.best_score_)
```

```python
forest_search = GridSearchCV(estimator=forest_model, param_grid=forest_param_grid,
                             scoring="balanced_accuracy", cv=cv, n_jobs=-1)

forest_search.fit(X_train, y_train)

best_forest_model = forest_search.best_estimator_

print("Best random forest parameters:", forest_search.best_params_)

print("Best random forest CV balanced accuracy:", forest_search.best_score_)
```

```python
from sklearn.model_selection import cross_val_predict

log_cv_prob = cross_val_predict(
    best_log_model, X_train, y_train, cv=cv, method="predict_proba")[:, 1]

forest_cv_prob = cross_val_predict(
    best_forest_model, X_train, y_train, cv=cv, method="predict_proba")[:, 1]
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score, balanced_accuracy_score
import numpy as np


def choose_threshold(y_true, probabilities):
    threshold_results = []

    for threshold in np.arange(0.10, 0.91, 0.05):
        predictions = (probabilities >= threshold).astype(int)
        boom_recall = recall_score(
            y_true, predictions, pos_label=1, zero_division=0)
        no_boom_recall = recall_score(
            y_true, predictions, pos_label=0, zero_division=0)

        threshold_results.append({"threshold": threshold, 'balanced_accuracy': balanced_accuracy_score(y_true, predictions),
                                  'boom_precision': precision_score(y_true, predictions, pos_label=1, zero_division=0), 'boom_recall': boom_recall,
                                  'no_boom_recall': no_boom_recall, 'macro_f1': f1_score(y_true, predictions, average='macro', zero_division=0)})

    threshold_df = pd.DataFrame(threshold_results)

    selected_threshold = (threshold_df.sort_values(
        ['balanced_accuracy', 'macro_f1'], ascending=False).iloc[0])

    return selected_threshold, threshold_df
```

```python
log_threshold_result, log_threshold_df = (
    choose_threshold(y_train, log_cv_prob))

forest_threshold_result, forest_threshold_df = (
    choose_threshold(y_train, forest_cv_prob))

log_threshold = float(log_threshold_result["threshold"])

forest_threshold = float(forest_threshold_result["threshold"])

print("Selected logistic threshold:", log_threshold)

display(log_threshold_result.to_frame().T)

print("Selected random forest threshold:", forest_threshold)

display(forest_threshold_result.to_frame().T)
```

```python
from sklearn.metrics import classification_report

log_cv_pred = (log_cv_prob >= log_threshold).astype(int)

forest_cv_pred = (forest_cv_prob >= forest_threshold).astype(int)

print("Logistic Regression: "
      "Cross-Validated Training Results")

print(classification_report(y_train, log_cv_pred, target_names=[
      "No boom",   "CONUS boom"], zero_division=0))

print("Random Forest: "
      "Cross-Validated Training Results")

print(classification_report(y_train, forest_cv_pred,
      target_names=["No boom",    "CONUS boom"], zero_division=0))
```

```python
log_test_prob = (best_log_model.predict_proba(X_test)[:, 1])

forest_test_prob = (best_forest_model.predict_proba(X_test)[:, 1])

log_test_pred = (log_test_prob >= log_threshold).astype(int)

forest_test_pred = (forest_test_prob >= forest_threshold).astype(int)
```
